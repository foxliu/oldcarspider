```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
mysql_url = 'mysql://test:123123@127.0.0.1:3306/spider'
engine = create_engine(mysql_url)
sql = 'select * from car_info;'
```

```python
data_frame = pd.read_sql(sql, engine)
```

### 把上牌时间序列更改为时间序列

```python
from dateutil.parser import parse
data_frame['shangpai_time'] = data_frame['shangpai_time'].map(parse)
```

```python
data_frame
```

```python
data_frame.groupby(data_frame['province'])['licheng'].mean()
```

```python
from datetime import datetime
data_frame[u'车辆年限'] = (datetime.now() - data_frame['shangpai_time']).map(lambda x: x.days/365)
```

```python
data_frame['price_diff'] = (data_frame['newcarprice'] - data_frame['price'])/data_frame['newcarprice']
```

```python
data_frame.sort_values(['province', '车辆年限'])
```

```python
price_diff = data_frame.groupby([data_frame['province'], data_frame['车辆年限']])['price_diff']
result_mean = price_diff.mean()
result_std = price_diff.std()
result_var = price_diff.var()
result_mean = result_mean.unstack().T
result_std = result_std.unstack().T
result_var = result_var.unstack().T
```

```python
max_city = result_mean.T.apply(lambda x: x.argmax())
max_num = result_mean.max(axis=1)
```

```python
result_mean['折价最多的省份'] = max_city
result_mean['折价最多的数字'] = max_num 
```
